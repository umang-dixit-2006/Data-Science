```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])
plt.title('Marks Scored by Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

```python
import matplotlib.pyplot as plt

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Highlight the region with the highest revenue
explode = [0.1 if r == max(revenue) else 0 for r in revenue]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode, shadow=True)
plt.title("Company Revenue Distribution by Region")
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random integers between 1 and 100
np.random.seed(0)
data = np.random.randint(1, 101, 1000)

plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of Random Integers (1-100)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
